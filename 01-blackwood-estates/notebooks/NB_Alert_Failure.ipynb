```python
# NB_Alert_Failure — logs pipeline failure
import notebookutils
from datetime import datetime
from pyspark.sql import Row
import uuid

#Use locals().get to safely capture the error message from the pipeline
error_msg = locals().get("error", "Unknown pipeline error")

# Create the failure log row
spark.createDataFrame([Row(
    run_id            = str(uuid.uuid4()),
    pipeline_name     = "PL_Blackwood_Wildcard",
    run_timestamp     = datetime.utcnow(),
    status            = "Failed",
    files_found       = 0,
    files_processed   = 0,
    files_quarantined = 0,
    rows_loaded       = 0,
    duration_seconds  = 0,
    error_message     = str(error_msg)[:500], # Safely truncate the error
    notes             = "Automatic failure log triggered"
)]) \
.write.format("delta").mode("append") \
.saveAsTable("Blackwood_Bronze_Lakehouse.pipeline_log")

print("Failure logged successfully")
```
